## 5. Feature Engineering and Modeling

### 5.1 Data Import

```python
# Imports:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from datetime import datetime
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LassoCV, RidgeCV
from sklearn import metrics
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
import statsmodels.api as sm
import warnings
```

```python
# Set max colums display to see all the columns
pd.options.display.max_columns = 80
```

```python
# Import and take a look at train_df
df = pd.read_csv("../data/train_filtered_1.csv")
df.head()
```

### 5.2 Features Engineering

```python
# Create age when the unit was sold
df["sold_age"] = df["tranc_year"] - df["lease_commence_date"]
df[["sold_age", "tranc_year", "lease_commence_date"]].head()
```

```python
# # THIS IS FOR STREAMLIT APPS
# # Drop others
# use_col = ['resale_price', 'planning_area', 'floor_area_sqft', 
#           'flat_type', 'sold_age', 
#           'mrt_nearest_distance', 'bus_stop_nearest_distance',
#           'mall_nearest_distance', 'hawker_nearest_distance',
#           'pri_sch_nearest_distance', 'sec_sch_nearest_dist' 
#           ]

# df = df[use_col]
```

```python
X = df.iloc[:,2:]
X.head()
```

```python
y = df.iloc[:,:1]
y.head()
```

### 5.2 Scaler

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.columns
```

```python
# Separate categorical and numerical columns
cat_col = []
num_col = []
for col in X_train.columns:
    if set(X_train[col].unique()) == set([0, 1]) or type(X_train[col].unique()[0]) == str:
        cat_col.append(col)
    else:
        num_col.append(col)
```

```python
cat_col
```

```python
num_col
```

```python
# Use make_column_transformer for onehotencoder and standardscaler
ohe = OneHotEncoder(min_frequency = 10)
ss = StandardScaler()
ct = make_column_transformer(
    (ohe, cat_col),
    (ss, num_col)
)
```

```python
# Fit and transform
X_train_sc = ct.fit_transform(X_train)
X_test_sc = ct.transform(X_test)
```

```python
X_train_sc
```

### 5.2 Model

```python
# Linear regression cross val
lr = LinearRegression()
lr_scores = cross_val_score(lr,X_train_sc,y_train,cv=3)
lr_scores.mean()
```

```python
# Ridge cross val
ridge = Ridge()
ridge_scores = cross_val_score(ridge,X_train_sc,y_train,cv=3)
ridge_scores.mean()
```

```python
# Lasso cross val
lasso = Lasso()
lasso_scores = cross_val_score(lasso,X_train_sc,y_train,cv=3)
lasso_scores.mean()
```

### Linear Regression

```python
# Use Linear Regression because we got the highest R2 score.
lr.fit(X_train_sc,y_train)
y_train_pred = lr.predict(X_train_sc)
y_test_pred = lr.predict(X_test_sc)

r2_train = metrics.r2_score(y_train, y_train_pred)
r2_test = metrics.r2_score(y_test, y_test_pred)

print(f"Trainig R2: {r2_train}")
print(f"Test R2: {r2_test}")
```

```python
# MSE 
metrics.mean_squared_error(y_test, y_test_pred)
```

```python
# RMSE
metrics.mean_squared_error(y_test, y_test_pred, squared=False)
```

```python
# Plot pred with residuals
residuals = y_test - y_test_pred
plt.scatter(y_test_pred, residuals, alpha = 0.25)
```
